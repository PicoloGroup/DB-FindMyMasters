```python
import pandas as pd
```

```python
masters = pd.read_csv("./masters_final.csv")
masters.head()
```

```python
masters.info()
```

```python

# to import postgres we need to remove NaN values
# also check enum values for mode, schedule, and field_of_study
# convert all double fields to int
```

```python

masters_clean = masters.fillna("")
masters_clean.head()
```

```python
def convert_to_int(col: str) -> str:
    """
    
    Convert a string value in a columns to int
    
    """
    
    if col == "":
        return col
    
    return int(col)
```

```python
masters_clean["duration"]      = masters_clean["duration"].apply(convert_to_int)
masters_clean["tution_amount"] = masters_clean["tution_amount"].apply(convert_to_int)
masters_clean["university_id"] = masters_clean["university_id"].apply(convert_to_int)

masters_clean.head()
```

```python
masters_clean["mode"].value_counts()
```

```python

# convert 'ONLINE_AND_CAMPUS COMBINED' to 'COMBINED'
def clean_mode(col: str) -> str:
    """
    
    Convert ONLINE_AND_CAMPUS COMBINED to COMBINED
    
    """
    
    if col == "ONLINE_AND_CAMPUS COMBINED":
        return "COMBINED"
    else:
        return col
```

```python
masters_clean["mode"] = masters_clean["mode"].apply(clean_mode)
```

```python
masters_clean.head()
```

```python
masters_clean["schedule"].value_counts()
```

```python

# rename field_of_study to field
masters_clean = masters_clean.rename(columns={'field_of_study': 'field'})
masters_clean.head()
```

```python

# finally match columns names for postgres
postgres_columns = ["id", "name", "language", "mode", "schedule", "deadline", "field", "url", "tution_currency", "university_id", "duration", "tution_amount"]

masters_pg = masters_clean[postgres_columns]
masters_pg.head()
```

```python
for field in masters_pg["field"].unique():
    print(field)
```

```python
masters_pg.to_csv("masters_pg.csv", index=False)
```
