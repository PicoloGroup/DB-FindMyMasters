```python
cities = pd.read_csv("../data-version2/raw/cities/city-quality-of-indexes.csv")
cities.head()
```

```python
cities.info()
```

```python
# for null values we have '?' in table we need to replace it with NULL for database

cities_clean = cities.replace('?', 'NULL')
```

```python
cities_clean.head()
```

```python
# rename columns
columns_name = {column: column.replace(' ', "_").lower() for column in cities.columns}
```

```python
columns_name
```

```python
cities_clean.rename(columns=columns_name, inplace=True, )
```

```python
cities_clean.head()
```

```python
# we will not use category data in postgress database

columns_to_postgres = [column for column in cities_clean.columns if "_category" not in column]
```

```python
columns_to_postgres
```

```python
cities_clean = cities_clean[columns_to_postgres]
```

```python
cities_clean.head()
```

```python
# need to reset index to have ids for each city

cities_clean.reset_index(inplace=True)
cities_clean.head()
```

```python
cities_clean.rename(columns={"index": "id"}, inplace=True)
cities_clean.head()
```

```python
# save cities csv

cities_clean.to_csv("./clean/cities_clean.csv", index=False)
```
