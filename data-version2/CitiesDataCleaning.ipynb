```python
cities = pd.read_csv("../data-version2/raw/cities/city-quality-of-indexes.csv")
cities.head()
```

```python
cities.info()
```

```python
# for null values we have '?' in table we need to replace it with NULL for database

cities_clean = cities.replace('?', 'NULL')
```

```python
cities_clean.head()
```

```python
# rename columns
columns_name = {column: column.replace(' ', "_").lower() for column in cities.columns}
```

```python
columns_name
```

```python
cities_clean.rename(columns=columns_name, inplace=True, )
```

```python
cities_clean.head()
```

```python
# we will not use category data in postgress database

columns_to_postgres = [column for column in cities_clean.columns if "_category" not in column]
```

```python
columns_to_postgres
```

```python
cities_clean = cities_clean[columns_to_postgres]
```

```python
cities_clean.head()
```

```python

# to clean name of cities and countries

import re

def clean_name(name: str) -> bool:
    """
    
    Simple check names. A valid name should only contain A-Z characters or space.
    
    """
    
    clean_name = re.sub('[^a-zA-Z ]', '', name)
    
    return " ".join(clean_name.split()).title()
```

```python
cities_clean["city"]    = cities_clean["city"].apply(clean_name)
cities_clean["country"] = cities_clean["country"].apply(clean_name)
```

```python
cities_clean.head()
```

```python
# need to reset index to have ids for each city

cities_clean.reset_index(inplace=True)
cities_clean.head()
```

```python
cities_clean = cities_clean.rename(columns={"index": "id"})
cities_clean.head()
```

```python

reordered_columns = ["id", "city", "country", "quality_of_life_index", "purchasing_power_index",
                     "safety_index", "health_care_index", "cost_of_living_index", 
                     "property_price_to_income_ratio", "traffic_commute_time_index", "pollution_index",
                     "climate_index"]
```

```python
cities_clean = cities_clean[reordered_columns]
```

```python
cities_clean = cities_clean.rename(columns={"city": "name"})
cities_clean.head()
```

```python
# save cities csv

cities_clean.to_csv("./clean/cities_clean.csv", index=False)
```
