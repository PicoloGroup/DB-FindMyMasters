```python
masters_prev = masters_prev.fillna("NULL")
```

```python

# here a dummiest idea. we will find university from name and get id from finalized universities data
# there can be many ideal and good solution but not enough time sorry to me. 

final_universities = pd.read_csv("./universities.csv")
final_universities.head(10)
```

```python
final_universities.fillna("NULL", inplace=True)
```

```python
# convert university_images csv to dict 
# it will make easier to find university_images url

university_dict = {}

for index in final_universities.index:
    university_dict[final_universities.loc[index, "name"]] = final_universities.loc[index, "id"]

for key in list(university_dict.keys())[:3]:
    print(f"{key} = {university_dict[key]}")
    print()
```

```python
masters_prev_dict = {}

for index in masters_prev.index:
    masters_prev_dict[masters_prev.loc[index, "name"]] = masters_prev.loc[index, "university"]

for key in list(masters_prev_dict.keys())[:3]:
    print(f"{key} = {masters_prev_dict[key]}")
    print()
```

```python
import re

# fist clean masters and universities name than check similarity to match 
def clean_name(name: str) -> bool:
    """
    
    Simple check names. A valid name should only contain A-Z characters or space.
    
    """
    if name == "NULL":
        return name
    
    clean_name = re.sub('[^a-zA-Z ]', '', name)
    
    return " ".join(clean_name.split()).title()
```

```python
masters_prev["name"]       = masters_prev["name"].apply(clean_name)
masters_prev["university"] = masters_prev["university"].apply(clean_name)
```

```python
import difflib

def is_similar(seq1, seq2):
    return difflib.SequenceMatcher(a=seq1.lower(), b=seq2.lower()).ratio() > 0.90


def get_university_id(row: str) -> str:
    """
    
    Get city id for university
    Because name may not match fully we will use similarity function
    
    :param name: row from university table
    :return    : city id if we find
    
    
    """
    
    master_id   = row["id"]
    master_name = row["name"]
            
    if master_name == "NULL":
        return "NULL"
    
    university_name: str = "NULL"
    
    # find univesity name first
    for master in masters_prev_dict.keys():
        
        if master == "NULL":
            continue
        
        if is_similar(master, master_name):
            university_name = masters_prev_dict[master]
            break
            
    if university_name == "NULL":
        return "NULL"
            
    for university in university_dict.keys():
                
        if is_similar(university, university_name):
            
            # keep university id
            university_id = university_dict[university]
            print(f"{master_id}/34539")
            
            return university_id
    
    
        
    return "NULL"
```

```python
masters["university_id"] = masters.apply(get_university_id, axis=1)
```
